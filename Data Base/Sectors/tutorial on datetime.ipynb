```python
import pandas as pd
import numpy as np
import datetime as dt
import yfinance as yf
import fredapi as fa
import requests
import json
import prettytable
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from pandas.tseries.frequencies import to_offset
```

```python
def data_grabber( ticker):
        #Function designed to take any ticker and grab annual data
        ticker = yf.Ticker(ticker)
        ticker_data = ticker.history(period = "max", interval = "1d")
        
        if not ticker_data.empty:
            return ticker_data
        else:
            print(f"The associated data for {ticker} is empty")
       
        return ticker_data
    
    #Overall Market Metrics
```

```python
df_AMZN = data_grabber("AMZN")
```

```python
display(df_AMZN.iloc[-61:-1])
```

```python
df_AMZN.index = pd.to_datetime(df_AMZN.index)
```

```python
weekly_df = df_AMZN.resample('W').first()
weekly_df.index = df_AMZN.groupby(df_AMZN.index.to_period('W')).apply(lambda x: x.index.min()).values
```

```python
display(weekly_df)
```

```python
monthly_df = df_AMZN.resample('M').first()
monthly_df.index = df_AMZN.groupby(df_AMZN.index.to_period('M')).apply(lambda x: x.index.min()).values
```

```python
display(monthly_df)
```

```python
weekly_returns = (weekly_df["Close"].pct_change()).dropna()
monthly = (monthly_df["Close"].pct_change()).dropna()
```

```python
display(weekly_returns)
display(monthly)
```
